Traing custom GrALE-like model and the Graph -> LLM projector

```
!python prototype/graph_llm_aligned.py -f prototype/grale_knn/grale_knn_llm_sft \
    -base_llm Qwen/Qwen2.5-1.5B-Instruct \
    -do_grale \
    -grale_batch_size 16 \
    -grale_epochs 2 \
    -do_proj \
    -proj_batch_size 8 \
    -proj_epochs 1
```

Fine-tune the LLM with SFT + PPO

```
!python prototype/knn_tanimoto_grale_generate.py -f_data data_baseline/data -f prototype/grale_knn/ -f_grale prototype/grale_knn/grale_knn_llm_sft \
  --do_finetune \
  --do_sft \
  --k 5 \
  --use_lora \
  --sft_steps 500 \
  --sft_lora 128 \
  --per_device_bs 1 \
  --grad_accum 8 \
  --base_llm grale \
  --out_llm_dir grale_knn_llm \
  --max_prompt_length 2048
```

Generate on validation with the fine-tuned model + evaluation metrics

```
!python prototype/knn_tanimoto_grale_generate.py -f_data data_baseline/data -f prototype/grale_knn/ \
  --do_generate --split validation --k 5 \
  --gen_batch_size 8 \
  --base_llm grale \
  --out_llm_dir grale_knn_llm_sft \
  --max_prompt_length 2048
```
