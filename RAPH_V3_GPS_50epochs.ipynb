```
import pickle
import numpy as np
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
import evaluate

from data_utils import load_id2emb, load_descriptions_from_graphs, PreprocessedGraphDataset, collate_fn
from train_gcn_v3_gps import MolGNN

TRAIN_GRAPHS = f"{DATA}/train_graphs.pkl"
VAL_GRAPHS   = f"{DATA}/validation_graphs.pkl"
TRAIN_EMB_CSV= f"{DATA}/train_embeddings.csv"
MODEL_PATH   = f"{DATA}/model_checkpoint_v3_gps.pt"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("DEVICE:", DEVICE)

@torch.no_grad()
def predict_val_retrieval():
    # Train text embeddings (targets)
    train_emb = load_id2emb(TRAIN_EMB_CSV)
    train_ids = list(train_emb.keys())
    train_embs = torch.stack([train_emb[i] for i in train_ids]).to(DEVICE)
    train_embs = F.normalize(train_embs, dim=-1)

    # Train id -> description (to output)
    train_id2desc = load_descriptions_from_graphs(TRAIN_GRAPHS)

    # Load model
    emb_dim = train_embs.shape[1]
    model = MolGNN(out_dim=emb_dim).to(DEVICE)
    state = torch.load(MODEL_PATH, map_location=DEVICE)
    model.load_state_dict(state)
    model.eval()

    # Validation graphs loader (only graphs)
    val_ds = PreprocessedGraphDataset(VAL_GRAPHS)
    val_dl = DataLoader(val_ds, batch_size=64, shuffle=False, collate_fn=collate_fn)

    # References (id -> true description)
    val_graphs = pickle.load(open(VAL_GRAPHS, "rb"))
    val_ref = {g.id: g.description for g in val_graphs}

    preds, refs = [], []
    offset = 0
    for batch in val_dl:
        batch = batch.to(DEVICE)
        mol_emb = model(batch)  # normalized in MolGNN
        sims = mol_emb @ train_embs.t()
        best = sims.argmax(dim=-1).cpu().tolist()

        bsz = batch.num_graphs
        batch_ids = val_ds.ids[offset:offset+bsz]
        offset += bsz

        for i, vid in enumerate(batch_ids):
            tid = train_ids[best[i]]
            preds.append(train_id2desc[tid])
            refs.append(val_ref[vid])

    return preds, refs

def compute_metrics(preds, refs):
    bleu = evaluate.load("bleu").compute(
        predictions=preds,
        references=[[r] for r in refs],
        max_order=4
    )["bleu"]  # parfois 0..1, parfois 0..100 selon version

    bert = evaluate.load("bertscore").compute(
        predictions=preds,
        references=refs,
        model_type="roberta-base",
        lang="en"
    )
    bert_f1 = float(np.mean(bert["f1"]))

    bleu_norm = bleu/100.0 if bleu > 1.0 else bleu
    final_proxy = 0.5*bleu_norm + 0.5*bert_f1
    return bleu, bert_f1, final_proxy

preds, refs = predict_val_retrieval()
bleu4, bert_f1, final_proxy = compute_metrics(preds, refs)

print("BLEU-4:", bleu4)
print("BERTScore F1:", bert_f1)
print("Final proxy (avg):", final_proxy)

print("\nExample:")
print("PRED:", preds[0][:200])
print("REF :", refs[0][:200])
```
