Fine-tune the LLM with GRPO (prototype run)

```python
!python knn_generate.py -f_data data_baseline/data -f prototype/knn/ \
  --do_finetune \
  --k 5 \
  --steps 200 \
  --max_train_samples 300 \
  --use_lora \
  --per_device_bs 2 --grad_accum 4 \
  --num_generations 4 \
  --bert_reward_bs 16 \
  --base_llm Qwen/Qwen2-0.5B-Instruct \
  --out_llm_dir knn_llm_grpo
```

```python
!python knn_generate.py -f_data data_baseline/data -f prototype/knn/ \
  --do_generate --split validation --k 5 \
  --gen_batch_size 8 --max_new_tokens 160 \
  --out_llm_dir knn_llm_grpo
```
