```
%%writefile /content/BornToOverfit/Methode3/run_benchmark.sh
set -e

DATA_DIR="/content/BornToOverfit/Methode3/data"
OUT_DIR="/content/BornToOverfit/Methode3/results"
SUB_DIR="$OUT_DIR/submissions"
CKPT_DIR="$OUT_DIR/checkpoints"
mkdir -p "$OUT_DIR" "$SUB_DIR" "$CKPT_DIR"

echo "encoder,val_bleu4,ckpt,test_csv" > "$OUT_DIR/benchmark.csv"

run_one () {
  ENC_NAME="$1"
  TRAIN_EMB="$2"
  VAL_EMB="$3"
  CKPT="$CKPT_DIR/model_${ENC_NAME}.pt"
  TESTCSV="$SUB_DIR/test_${ENC_NAME}.csv"

  echo "============================"
  echo "Training: $ENC_NAME"
  echo "============================"

  python /content/BornToOverfit/Methode3/train_gcn_v3_gps.py \
    --data_dir "$DATA_DIR" \
    --train_emb "$TRAIN_EMB" \
    --val_emb "$VAL_EMB" \
    --epochs 50 \
    --out_ckpt "$CKPT"

  echo "Eval val (BLEU4): $ENC_NAME"
  BLEU=$(python /content/BornToOverfit/Methode3/eval_val_retrieval.py \
    --data_dir "$DATA_DIR" \
    --train_emb "$TRAIN_EMB" \
    --ckpt "$CKPT" | tail -n 1)

  echo "Generate test submission: $ENC_NAME"
  python /content/BornToOverfit/Methode3/retrieval_answer.py \
    --train_graphs "$DATA_DIR/train_graphs.pkl" \
    --test_graphs  "$DATA_DIR/test_graphs.pkl" \
    --train_emb_csv "$TRAIN_EMB" \
    --ckpt "$CKPT" \
    --output_csv "$TESTCSV"

  echo "$ENC_NAME,$BLEU,$CKPT,$TESTCSV" >> "$OUT_DIR/benchmark.csv"
  echo "✅ $ENC_NAME done. BLEU4=$BLEU"
}

# === 4 encoders ===
run_one "bert_base" \
  "$DATA_DIR/train_embeddings.google-bert_bert-base-uncased.cls.csv" \
  "$DATA_DIR/validation_embeddings.google-bert_bert-base-uncased.cls.csv"

run_one "scibert" \
  "$DATA_DIR/train_embeddings.allenai_scibert_scivocab_uncased.cls.csv" \
  "$DATA_DIR/validation_embeddings.allenai_scibert_scivocab_uncased.cls.csv"

run_one "pubmedbert" \
  "$DATA_DIR/train_embeddings.microsoft_BiomedNLP-BiomedBERT-base-uncased-abstract.cls.csv" \
  "$DATA_DIR/validation_embeddings.microsoft_BiomedNLP-BiomedBERT-base-uncased-abstract.cls.csv"

run_one "chemical_bert" \
  "$DATA_DIR/train_embeddings.recobo_chemical-bert-uncased.cls.csv" \
  "$DATA_DIR/validation_embeddings.recobo_chemical-bert-uncased.cls.csv"

echo "✅ DONE"
echo "Benchmark saved to: $OUT_DIR/benchmark.csv"
```

```
!pip -q install bert-score nltk transformers tqdm

from google.colab import drive
drive.mount("/content/drive")

import os, re, pickle, math
import pandas as pd
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModel
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

# -------------------------
# 0) EDIT THIS
# -------------------------
RUN_RESULTS_DIR = "/content/drive/MyDrive/ALTEGRAD/benchmarks/methode3_textenc_20251231_020406/results"
CKPT_DIR = os.path.join(RUN_RESULTS_DIR, "checkpoints")
assert os.path.exists(CKPT_DIR), f"Missing CKPT_DIR: {CKPT_DIR}"

# -------------------------
# 1) Find graphs (Drive baseline preferred)
# -------------------------
CANDIDATE_DATA_DIRS = [
    "/content/drive/MyDrive/ALTEGRAD/data_baseline/data",
    "/content/BornToOverfit/Methode3/data",
]
DATA_DIR = None
for d in CANDIDATE_DATA_DIRS:
    if os.path.exists(os.path.join(d, "train_graphs.pkl")) and os.path.exists(os.path.join(d, "validation_graphs.pkl")):
        DATA_DIR = d
        break
assert DATA_DIR is not None, "Could not find train_graphs.pkl/validation_graphs.pkl."
print("✅ Using DATA_DIR:", DATA_DIR)
print("✅ Using CKPT_DIR:", CKPT_DIR)

TRAIN_GRAPHS = os.path.join(DATA_DIR, "train_graphs.pkl")
VAL_GRAPHS   = os.path.join(DATA_DIR, "validation_graphs.pkl")

# -------------------------
# 2) Import project code (MolGNN + dataset)
# -------------------------
import sys
for p in ["/content/BornToOverfit/Methode3"]:
    if os.path.exists(p) and p not in sys.path:
        sys.path.append(p)

from data_utils import PreprocessedGraphDataset, collate_fn

try:
    from train_gcn_v3_gps import MolGNN
    print("✅ Imported MolGNN from train_gcn_v3_gps.py")
except Exception:
    from train_gcn import MolGNN
    print("✅ Imported MolGNN from train_gcn.py")

# -------------------------
# 3) Helpers
# -------------------------
def load_graphs(pkl_path):
    with open(pkl_path, "rb") as f:
        return pickle.load(f)

def id2desc_from_graphs(graphs):
    return {str(g.id): (g.description or "") for g in graphs}

def mean_pool(last_hidden_state, attention_mask):
    mask = attention_mask.unsqueeze(-1).type_as(last_hidden_state)
    summed = (last_hidden_state * mask).sum(dim=1)
    denom = mask.sum(dim=1).clamp(min=1e-6)
    return summed / denom

@torch.no_grad()
def encode_text_batch(texts, tokenizer, model, device, max_len=128, pooling="cls"):
    tok = tokenizer(
        texts, return_tensors="pt",
        padding=True, truncation=True, max_length=max_len
    ).to(device)
    out = model(**tok)
    if pooling == "cls":
        emb = out.last_hidden_state[:, 0, :]
    else:
        emb = mean_pool(out.last_hidden_state, tok["attention_mask"])
    return emb.detach()

def simple_tokenize(s: str):
    s = (s or "").lower().strip()
    s = re.sub(r"[^a-z0-9\s]+", " ", s)
    s = re.sub(r"\s+", " ", s)
    return s.split()

def bleu4(preds, refs):
    smoothie = SmoothingFunction().method1
    list_of_refs = [[simple_tokenize(r)] for r in refs]
    hyps = [simple_tokenize(p) for p in preds]
    return float(corpus_bleu(list_of_refs, hyps, weights=(0.25,0.25,0.25,0.25), smoothing_function=smoothie))

def bertscore_f1(preds, refs, device):
    from bert_score import score as bertscore
    P, R, F1 = bertscore(
        cands=preds, refs=refs,
        lang="en",
        model_type="roberta-large",
        device=device,
        rescale_with_baseline=True,
        verbose=False,
    )
    return float(F1.mean().item())

# -------------------------
# 4) Build train text embedding matrix ON THE FLY
# -------------------------
@torch.no_grad()
def build_train_text_matrix(train_graphs_pkl, hf_model_name, device, pooling="cls", batch_size=64, max_len=128):
    graphs = load_graphs(train_graphs_pkl)
    ids = [str(g.id) for g in graphs]
    descs = [(g.description or "") for g in graphs]

    tok = AutoTokenizer.from_pretrained(hf_model_name)
    mdl = AutoModel.from_pretrained(hf_model_name).to(device)
    mdl.eval()

    embs = []
    for i in tqdm(range(0, len(descs), batch_size), desc=f"Encode train texts ({hf_model_name})"):
        batch = descs[i:i+batch_size]
        e = encode_text_batch(batch, tok, mdl, device, max_len=max_len, pooling=pooling)
        embs.append(e)

    mat = torch.cat(embs, dim=0)
    mat = F.normalize(mat, dim=-1)
    return ids, mat  # (N,), (N,H)

# -------------------------
# 5) Retrieve on VAL then compute BLEU4 + BERTScore
# -------------------------
@torch.no_grad()
def eval_model(tag, hf_model_name, ckpt_path, device):
    # Build train embedding matrix from HF encoder
    train_ids, train_mat = build_train_text_matrix(TRAIN_GRAPHS, hf_model_name, device=device, pooling="cls")

    # Load val graphs and references
    train_graphs = load_graphs(TRAIN_GRAPHS)
    val_graphs   = load_graphs(VAL_GRAPHS)
    train_id2desc = id2desc_from_graphs(train_graphs)
    val_id2desc   = id2desc_from_graphs(val_graphs)

    # Mol encoder output dim must match text emb dim
    out_dim = train_mat.size(1)
    model = MolGNN(out_dim=out_dim).to(device)
    model.load_state_dict(torch.load(ckpt_path, map_location=device))
    model.eval()

    # Encode val molecules
    val_ds = PreprocessedGraphDataset(VAL_GRAPHS)
    val_dl = DataLoader(val_ds, batch_size=64, shuffle=False, collate_fn=collate_fn)

    mol_vecs = []
    val_ids_ordered = []
    seen = 0
    for graphs in tqdm(val_dl, desc=f"Encode val mols ({tag})"):
        graphs = graphs.to(device)
        v = F.normalize(model(graphs), dim=-1)
        mol_vecs.append(v)
        bs = graphs.num_graphs
        val_ids_ordered.extend(val_ds.ids[seen:seen+bs])
        seen += bs

    val_mat = torch.cat(mol_vecs, dim=0)

    # Retrieve nearest train text
    sims = val_mat @ train_mat.t()
    best = sims.argmax(dim=-1).cpu().tolist()

    preds, refs, ids = [], [], []
    for i, vid in enumerate(val_ids_ordered):
        tid = train_ids[best[i]]
        preds.append(train_id2desc.get(tid, ""))
        refs.append(val_id2desc.get(str(vid), ""))
        ids.append(str(vid))

    # Metrics
    b4 = bleu4(preds, refs)
    bf1 = bertscore_f1(preds, refs, device=device)

    # Save per-model val predictions
    pred_csv = os.path.join(RUN_RESULTS_DIR, f"val_predictions_{tag}.csv")
    pd.DataFrame({"ID": ids, "pred": preds, "ref": refs}).to_csv(pred_csv, index=False)

    return {"model": tag, "BLEU4_val": b4, "BERTScoreF1_val": bf1, "val_pred_csv": pred_csv}

# -------------------------
# 6) Config of your 4 encoders (HF names) + checkpoints
# -------------------------
MODELS = {
    "bert_base":     "google-bert/bert-base-uncased",
    "scibert":       "allenai/scibert_scivocab_uncased",
    "pubmedbert":    "microsoft/BiomedNLP-BiomedBERT-base-uncased-abstract",
    "chemical_bert": "recobo/chemical-bert-uncased",
}

device = "cuda" if torch.cuda.is_available() else "cpu"
rows = []

for tag, hf_name in MODELS.items():
    ckpt = os.path.join(CKPT_DIR, f"model_{tag}.pt")
    assert os.path.exists(ckpt), f"Missing checkpoint: {ckpt}"
    print(f"\n====================\nEVAL: {tag}\nHF: {hf_name}\nCKPT: {ckpt}\n====================")
    rows.append(eval_model(tag, hf_name, ckpt, device))

df = pd.DataFrame(rows).sort_values("BERTScoreF1_val", ascending=False)
print("\n✅ SUMMARY (sorted by BERTScoreF1_val):")
display(df)

out_csv = os.path.join(RUN_RESULTS_DIR, "scores_textencoders_recomputed.csv")
df.to_csv(out_csv, index=False)
print("\n✅ Saved:", out_csv)
```
